```python
import matplotlib.pyplot as plt

#create data for plotting
x_values = [0,1,2,3,4,5]
y_values = [0,1,4,9,16,25]

#the default graph style for plot is a line
plt.plot(x_values,y_values)

#display the graph
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.xlabel('some numbers')
plt.show()
```

```python
import matplotlib.pyplot as plt

values = [5, 6, 3, 7, 2]
names = ["A", "B", "C", "D", "E"]

plt.bar(names, values, color = "green")

plt.show()
```

```python
import matplotlib.pyplot as plt

ages = [2,5,70,40,30,65,18,43]
range = (0, 100)
bins = 10

plt.hist(ages, bins, range, color='green', histtype='bar', edgecolor='black')

plt.xlabel('age')

plt.ylabel('No. of people')

plt.title('My Histogram')
```

```python
import matplotlib.pyplot as plt

x_values = [0,1,2,3,4,5]
y_values = [0,1,4,9,16,25]

plt.scatter(x_values, y_values, s=100, color="red", edgecolor="yellow", marker="*", label="stars")
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt

activities = ['eat', 'sleep', 'work', 'play']

slices = [3, 7, 8, 6]

colors = ['r', 'y', 'g', 'b']

plt.pie(slices, labels=activities, colors=colors, startangle=90, shadow=True, explode = (0, 0, 0.1, 0), radius = 1.2, autopct='%1.1f%%')

plt.legend()
```
